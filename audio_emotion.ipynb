```python
import os
import librosa
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

DATA_PATH = r"C:\Users\prath\face emotion\audio_speech_actors_01-24"

X = []
y = []

emotion_map = {
    "01": "neutral",
    "03": "happy",
    "04": "sad",
    "05": "angry"
}

for file in os.listdir(DATA_PATH):
    if file.endswith(".wav"):
        emotion_code = file.split("-")[2]

        if emotion_code in emotion_map:
            emotion = emotion_map[emotion_code]

            path = os.path.join(DATA_PATH, file)

            audio, sr = librosa.load(path, duration=3)
            mfcc = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=40)
            mfcc = np.mean(mfcc.T, axis=0)

            X.append(mfcc)
            y.append(list(emotion_map.values()).index(emotion))

X = np.array(X)
y = np.array(y)

# MODEL
model = Sequential()
model.add(Dense(128, activation='relu', input_shape=(40,)))
model.add(Dense(64, activation='relu'))
model.add(Dense(4, activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X, y, epochs=50)
```

```python
model.save("speech_model.h5")
```
